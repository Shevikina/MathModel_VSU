```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge
x1=np.arange(1,11)
y1=[-10.1,-8.3,-2.4,5.1,13.8,17.9,20,19.1,11.8,5]
plt.plot(x1,y1,label=r"Средняя температура в Казани")
plt.grid(True)
plt.legend()
plt.show()
```

```python
X=np.vstack([x1,np.ones(len(x1))])
X=X.T
lsq=np.linalg.lstsq(X,y1,rcond=None)
a,b=lsq[0]
```

```python
plt.plot(x1,a*x1+b)
plt.plot(x1,y1,".")
```

```python
def f(x1,a,b):
    return a*x1*x1+b*x1;
(a,b),_=curve_fit(f,x1,y1)
plt.plot(x1,f(x1,a,b))
plt.plot(x1,y1,".")
```

```python
x_m=x1[:,np.newaxis]
model=make_pipeline(PolynomialFeatures(1),Ridge())
model.fit(x_m,y1)
```

```python
x_long=np.linspace(1,12,100)[:,np.newaxis]
y_p=model.predict(x_long)
plt.plot(x_long,y_p)
plt.plot(x1,y1,".")
```
