### Мягкая модель Мальтуса с зависящим от населения коэффициентом размножения

$x'=k(x)*x$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dydt(y, t):
        a=0.5
        b=0.2
        return a*y-b*y*y
```

```python
t=np.linspace(0,10,100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y' = ay-by^2$")
plt.grid(True)
plt.legend()
plt.show()
```

### Жесткая модель рыболовства с интенсивностью c и количеством рыб x

$x'=x-x^2-c$

```python
def dxdt(x,t):
    c=0.21
    return x-x*x-c
t=np.linspace(0,10,100)
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
```

```python
%matplotlib inline
plt.plot(t,solve,label=r"$x'=x-x^2-c$")
plt.grid(True)
plt.legend()
plt.show()
```

### Мягкая модель рыболовства с критическим значением в виде максимума функции g(x)

$x'=x-g(x)*x-c$

```python
def dydt(y,t):
    a=0.8
    b=0.4
    c=0.3
    return y-(a-b*y)*y-c
```

```python
t=np.linspace(0,10,100)
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y' = y-(a-b*y)*y-c$")
plt.grid(True)
plt.legend()
plt.show()
```

### Модель изменения популяции рыб при дифференциальной квоте вылова

$x'=x-x^2-kx$

```python
def dxdt(x,t):
    k=4
    return x-x**2-k*x
t=np.linspace(0,10,100)
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
```

```python
%matplotlib inline
plt.plot(t,solve,label=r"$x'=x-x^2-kx$")
plt.grid(True)
plt.legend()
plt.show()
```

### Модель Мальтиуса роста населения Земли

$x'=-kx$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
def dxdt(x,t):
    k=0.21
    return -k*x
t=np.linspace(0,10,100)
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
```

```python
%matplotlib inline
plt.plot(t,solve,label=r"$x'=-kx$")
plt.grid(True)
plt.legend()
plt.show()
```
