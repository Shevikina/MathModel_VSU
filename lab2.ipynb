### Модель Мальтиуса роста населения Земли

$𝑥′=−𝑘𝑥$

```python
def dxdt(x,t):
    k=-0.031
    return -k*x
t=years
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
%matplotlib inline
plt.plot(years, people, c = 'g')
plt.plot(t,solve,label=r"$x'=-kx$")
plt.grid(True)
plt.legend()
plt.show()
```

### Мягкая модель Мальтуса с зависящим от населения коэффициентом размножения
$𝑥′=𝑘(𝑥)*𝑥$

```python
def dydt(y, t):
#         a=0.003
#         b=-0.0006
        a=0.001
        b=-0.0016
        return a*y-b*y*y
t=years
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label = r"$y' = ay-by^2$")
plt.plot(years, people, c = 'g')
plt.grid(True)
plt.legend()
plt.show()
```
